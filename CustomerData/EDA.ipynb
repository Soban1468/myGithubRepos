### Importing Libraries

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

### Loading Dataset

```python
df = pd.read_csv(r'F:\Python codes\machine-learning Practices\CustomerData\clean_customers.csv')
```

```python
# Dropping unnecessary columns
# 'customer_name' and 'customer_id' are not needed for analysis

df.drop(columns=['customer_name', 'customer_id'], inplace=True)
```

```python
df.head()
```

### Plotting Churn vs Non-Churn Graph

```python
ax = sns.countplot(x='churn' , data=df , palette = 'Set2')
plt.title("Churn vs Non-Churn customers")
plt.xlabel("Churn")
plt.ylabel("Customer Count")
ax.set_xticklabels(["Non-Churn (0)", "Churn (1)"])
plt.show()
```

### Displaying Correlation Heatmaps between Features

```python
plt.figure(figsize=(10, 5))
sns.heatmap(data=df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

### Checking top related Features with Churn

```python
corr = df.corr(numeric_only=True)['churn'].sort_values(ascending=False)
corr.head(6)
```

## Note: ##
Subscription status (Inactive/Active) and customer usage history appear to be more predictive of churn compared to demographic features like age or gender. Location also shows small but notable effects, with Toronto and New York customers having higher churn risk
